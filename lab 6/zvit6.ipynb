1. Робота у віртуальному середовищі
    1. Створив віртуальне середовище `VENV`
        - команди
            ```bash
            python -m venv ./my_env
            source my_env/Scripts/activate
            pip install requests
            deactivate
            pip show requests
            ```    
        - результат виконання `pip show requests`
            ```text 
            WARNING: Skipping reqeusts as it is not installed.
            ```
    1. Команда `pip show requests` вивела помилку, тому що не знайшла встановлену бібліотеку, через то що ми інсталювали її в середені віртуального середовища, і її не видно у загальній системі;
