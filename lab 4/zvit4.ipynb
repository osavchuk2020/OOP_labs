### Виконання роботи
1. Створив Python файл `*.py` для виконнання прикладів.
    1. Ознайомився з основними типами даних, такими як text, number, list, dict, tuple, set
        ```python
        a = "lorem ipsum" #string
        b = 1 # number
        c = ["a", 1, 1.25, "text"] # List
        d = {"a": "text", "b": 1} # Dict
        e = ("a", ) # Tuple
        f = {"ss", } # Set
        ```
    1. За допомогою команди `print` вивів декілька констант:
        - код програми 
        ```python
        print("constant FALSE = ", False)
        print("constant TRUE = ", True)
        print("constant NONE = ", None)
        ```
        - результат виконання 
        ```text
        constant FALSE =  False
        constant TRUE =  True
        constant NONE =  None
        ```
    1. Ознайомився з вбудованими функціями та вивів результат роботи деяких з них:
       - код програми 
        ```python
        print(abs(-12.5), f"equals {abs(12.5)}")
        print(pow(13, 2), f"equals {13**2}")
        print(round(10/3, 3))
        ```
        - результат виконання 
        ```text
        12.5 equals 12.5
        169 equals 169
        3.333
        ```
    1. Ознайомився з циклами, написав декілька прикладів їх роботи:
        - код програми 
        ```python
        letters = ["a", "b", "c"]
        print("list of letters:")
        for i in range(len(letters)):
            print(f"{i+1}. {letters[i]}")

        i = 0
        while letters[i] != "c":
            print(f"{i}. {letters[i]}")
            i += 1
        ```
        - результат виконання 
        ```text
        list of letters:
        1. a
        2. b
        3. c

        0. a
        1. b        
        ```
    1. Ознайомився з принципом роботи розгалуджень, виконав декілька прикладів їх роботи:
       - код програми 
        ```python
        print("b = 1" if b == 2 else f"b = {b}")
        print(f"name = {name}" if "name" in locals() else "variable `name` doesnt exist")
        ```
        - результат виконання 
        ```text
        b = 1
        variable `name` doesnt exist
        ```
    1. Приклад роботи конструкції `try`->`except`->`finally`для виявлення можливих помилок у коді:
       - код програми 
        ```python
        negNum = -100
        try: 
            print(f"square root of {negNum} equals", math.sqrt(negNum))
        except Exception as e:
            print("ERROR:", e)
        finally:
            print("(this math operation cannot be performed)")
        ```
        - результат виконання 
        ```text
        ERROR: math domain error
        (this math operation cannot be performed)
        ```
    1. Приклад роботи контекст-менеджеру `with`:
       - код програми 
        ```python
        with open("textFile.txt", 'w') as fileObject:
        fileObject.write("lorem ipsum")
        ```
        - в результаті виконання програми текст записується у файл
    1. Приклад роботи анонімних функцій `lambda`:
        ```python
        compare = lambda a, b: f"{a} > {b}" if a > b else f"{a} < {b}" if a < b else f"{a} = {b}"
        print(compare(3, 2))
        ```
